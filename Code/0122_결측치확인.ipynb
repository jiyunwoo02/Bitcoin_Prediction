```python
# 결측치가 있는 행 식별
missing_rows = df.isnull().sum(axis=1)
```

```python
# 결측치가 있는 행의 수 계산
num_missing_rows = len(missing_rows[missing_rows > 0])
```

```python
print("결측치가 있는 행의 수:", num_missing_rows)
```
