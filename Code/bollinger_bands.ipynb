```python
import pandas as pd
import numpy as np
```

```python
# 데이터 불러오기, 시계열데이터로 변환, 칼럼'time' 인덱스 설

df = pd.read_csv(r"C:\Users\joan\OneDrive\문서\donguri\비트코인 공모전\BIT_2023_1차.csv")
df.set_index('time', inplace=True)
df.index=pd.to_datetime(df.index)
df
```

```python
#볼린저 밴드 계산 함수
#파라미터 정의) window : 이동평균 계산 기준 일자, 20일로 설정(임의), num_std_dev : 표준편차 배수, 2로 설정(계산식 참고)
#참고) window = 20으로 첫 20개 행에서 'NaN' 값 출력됨. 해당 결측값 처리 방법 논의 필요 

def bollinger_bands(df, window=20, num_std_dev=2):
    
    # 이동평균 계산
    df['rolling_mean'] = df['price'].rolling(window=window).mean()

    # 표준편차 계산
    df['rolling_std'] = df['price'].rolling(window=window).std()

    # 볼린저 밴드 상단 및 하단 계산
    df['upper_band'] = df['rolling_mean'] + (num_std_dev * df['rolling_std'])
    df['lower_band'] = df['rolling_mean'] - (num_std_dev * df['rolling_std'])

    # 필요없는 컬럼 삭제
    df.drop(['rolling_mean', 'rolling_std'], axis=1, inplace=True)

    return df

# 볼린저 밴드 계산
df_bollinger_bands = bollinger_bands(df)
df_bollinger_bands
```

```python
# 필요한 컬럼만 남기고 drop (세희님 코드에 맞춰서 추가했습니다.)

df_bollinger_bands.drop(['Open', 'High', 'Low', 'Close', 'Volume', 'returns', 'volatility',
       'price', 'quote_qty', 'is_buyer_maker', 'quote_qty_cal'], axis=1, inplace=True)
df_bollinger_bands
```

```python
#csv 파일로 내보내기
df_bollinger_bands.to_csv('BIT_2023_bollinger_bands.csv')
```
